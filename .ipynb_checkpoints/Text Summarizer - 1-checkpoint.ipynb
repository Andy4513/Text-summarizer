## Get the category, url, title and summary of articles

```python
category_list = []
url_list = []
title_list = []
summary_list = []

ua = UserAgent()
header = {'User-Agent':str(ua.chrome)}

url = 'https://www.newsweek.com/newsfeed?page='

start_time = time()

for i in range(1, 51):  # Get 50 pages

  # Make a requuest
  response = requests.get(url + str(i), headers = header)

  # Pause the loop
  sleep(randint(5, 8))

  # Monitor request
  elapsed_time = time() - start_time
  print('Request: {}; Frequency: {} requests/s'.format(i, i/elapsed_time))
  

  # Throw a warning if code is not 200
  if response.status_code != 200:
    warnings.warn('Request: {}; Status code: {}'.format(i, response.status_code))

  # Get articles in page
  page_html = bs4.BeautifulSoup(response.text, 'html.parser') # 
  articles = page_html.find_all('article')
  
  # Loop through articles and get info
  for article in articles:
    category_list.append(article.find('div', class_ = 'category').text)
    url_list.append(article.find('h3').a['href'])
    title_list.append(article.find('h3').text)
    summary_list.append(article.find('div', class_ = 'summary').text)
```

```python
d = {
    'category': category_list,
    'url': url_list,
    'title': title_list,
    'summary': summary_list
}
df = pd.DataFrame(d)
df
```

```python
df.to_csv('news_data.csv')
```

## Get the text of the full article

```python
text_list = []

base_url = 'https://www.newsweek.com'

start_time = time()

request = 0

# Loop through all the url
for url in url_list:
  # Make a requuest
  response = requests.get(base_url + url, headers = header)
  request += 1

  # Pause the loop
  sleep(random())

  # Monitor request
  elapsed_time = time() - start_time
  print('Request: {}; Frequency: {} requests/s'.format(request, request/elapsed_time))
  if request % 10 == 0:
    output.clear()

  # Throw a warning if code is not 200
  if response.status_code != 200:
    warnings.warn('Request: {}; Status code: {}'.format(request, response.status_code))

  # Find div that contains the article text
  page_html = bs4.BeautifulSoup(response.text, 'html.parser') 
  div = page_html.find('div', class_ = 'article-body v_text')

  # Get the text
  if div != None:
    all_p = div.find_all('p')
    if all_p != None:
      text = "".join([p.text for p in all_p])
      text_list.append(text)
    else:
      text_list.append('')
  else:
    text_list.append('')
```

```python
df['full_text'] = text_list
df
```
