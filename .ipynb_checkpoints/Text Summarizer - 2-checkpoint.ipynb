# Summarize the whole data set

```python
text_summaries = []

for i in range(len(df)): # iterate through all rows of data frame
  # Get the full text for the current row
  full_text = df['full_text'][i] 

  if full_text is None:
    text_summaries.append("")
    continue

  # Tokenize the sentences and words
  sentence_tokens = nltk.tokenize.sent_tokenize(str(full_text))
  sentence_tokens_lower = [sentence.lower() for sentence in sentence_tokens]
  sentence_tokens_lower = [re.sub(r'[^a-zA-Z ]', ' ', sentence) for sentence in sentence_tokens_lower]
  word_tokens = [nltk.tokenize.word_tokenize(sentence) for sentence in sentence_tokens_lower]

  # Get words frequency
  words = []
  for word_list in word_tokens:
    for word in word_list:
      words.append(word)
  word_freq = Counter(words)

  # Calculate sentece weights based on words frequency
  sentence_weights = []
  for sentence in word_tokens:
    weight = 0
    for word in sentence:
      weight += word_freq[weight]
    sentence_weights.append(weight)
  
  # Get the indeces of the 2 sentences with the highest sentence weight and add to text_summaries
  try:
    index_1, index_2 = np.argsort(sentence_weights)[::-1][:2]
  except:
    text_summaries.append(sentence_weights[0])
  else:
    text_summaries.append(' '.join([sentence_tokens[index_1], sentence_tokens[index_2]]))


df['extracted_summary'] = text_summaries
df.head()
```

```python
df.to_csv('drive/MyDrive/Colab Notebooks/news_data_with_sum.csv')
```
